```python
def greet (name):
  return  f"Hello {name}, welcome!"
  print(greet("Rahul"))
```

```python
def introduction(name,age):
  return f"my name is{name}and i am {age} years old"
  print(introduction(age=26,name="rahul"))
```

```python
def power(base,exponent):
  return base**exponent
  print(power(2,3))
```

```python
def add_number(*args):
  total=0
  for num in args:
    total+=num
  return "sum is" + str(total)
print(add_number(10,20))
print(add_number(5,15,25,35))
```
